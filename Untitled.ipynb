```python
import numpy as np
import pandas as pd
import os 
import shutil
```

```python
import tensorflow as tf
```

```python
#creaet the data for positives samples(github)
file_path="chestxray/metadata.csv"
images_path="chestxray/images"

#
```

```python
df=pd.read_csv(file_path)
print(df.shape)
```

```python
target_dir="Dataset/covid"
if not os.path.exists(target_dir):
    os.mkdir(target_dir)
    print("created")
```

```python
#transferring
cnt=0
for (i,row) in df.iterrows():
    if row["finding"]=="Pneumonia/Viral/COVID-19" and row["view"]=="PA":
        filename=row["filename"]
        #full_path = os.path.join(folder, file)
        image=os.path.join(images_path,filename)
        image_copy=os.path.join(target_dir,filename)
        shutil.copy2(image,image_copy)
       # print("moving",cnt)
        cnt+=1
        
print(cnt)
```
